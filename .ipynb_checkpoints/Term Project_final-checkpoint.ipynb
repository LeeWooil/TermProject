```python
import folium
from folium.plugins import MarkerCluster, MiniMap

data = pd.read_csv("./SchoolZoneAcc_info.csv",encoding="euc-kr")


map_hs = folium.Map((37.470493, 127.057265), zoom_start = 10)
mc = MarkerCluster() 


Loc = list(data['loc'])
Occurs = list(data['occurs']) 
Casualties = list(data['caslt'])
Dead = list(data['dprs'])
Serious_injury = list(data['serinjry'])
Slight_injury = list(data['sltinjry'])
Latitude = list(data['lat'])
Longitude = list(data['logt'])

for i in range(len(Loc)):
    mc.add_child(folium.Marker(location = [Latitude[i], Longitude[i]], 
                               popup=[Loc[i],Occurs[i],Casualties[i],Dead[i],Serious_injury[i],
                                      Slight_injury[i]])) 
    
    
    map_hs.add_child(mc)

map_hs.save("./SchoolAcc1.html")
```

```python
def getSchoolZone_Info():
    
    url = "https://openapi.gg.go.kr/ChildProtectionArea?KEY=737c37d3f5784ef881b0f7e272c66018&Type=json"
    total_count=requests.get(url).json()['list_total_count']
    
    loc = []
    cctv_yn = []
    cctv_cnt = []
    road_bt = []
    latitude = []
    longitude = []
    
    for count in range(1,totla_count+1):
        Schoolzone_info = requests.get(url).json()['row']
        
        for info in Schoolzone_info:
            loc.append(info['TARGET_FACLT_NM'])
            cctv_yn.append(info['CCTV_INSTL_YN'])
            cctv_cnt.append(info['CCTV_INSTL_CNT'])
            road_bt.append(info['PRTCAREA_ROAD_BT_INFO'])
            latitude.append(info['REFINE_WGS84_LAT'])
            longitude.append(info['REFINE_WGS84_LOGT'])
            
    School_zone_info_df = pd.DataFrame({"TARGET_FACLT_NM":loc,"CCTV_INSTL_YN":cctv_yn
                                       ,"CCTV_INSTL_CNT":cctv_cnt,"PRTCAREA_ROAD_BT_INFO":road_bt,
                                      "REFINE_WGS84_LA":latitude, "REFINE_WGS84_LOGT":longitude })
    return School_zone_info_df

my_info_df = getSchoolZone_Info()
```

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests


SchoolZone_info = {}
pIndex = 1

loc_list = []
cctv_yn_list = []
cctv_cnt_list = []
road_bt_list = []
lat_list = []
logt_list = []

while pIndex<=5:
    url = "https://openapi.gg.go.kr/ChildProtectionArea?KEY=737c37d3f5784ef881b0f7e272c66018&Type=xml&'pIndex'&pSize=2"
    req = requests.get(url)
    html = req.text
    soup = BeautifulSoup(html,'html.parser')
    
    loc = soup.find_all('target_faclt_nm')
    cctv_yn = soup.find_all('cctv_instl_yn')
    cctv_cnt = soup.find_all('cctv_instl_cnt')
    road_bt = soup.find_all('prtcarea_road_bt_info')
    lat = soup.find_all('refine_wgs84_lat')
    logt = soup.find_all('refine_wgs84_logt')
                       
    for code in loc:
        loc_list.append(code.text)
    for code in cctv_yn:
        cctv_yn_list.append(code.text)
    for code in cctv_cnt:
        if cctv_cnt == "<cctv_instl_cnt></cctv_instl_cnt>":
            cctv_cnt_list.append(0)
        else :
            cctv_cnt_list.append(code.text)
    for code in road_bt:
        road_bt_list.append(code.text)
    for code in lat:
        lat_list.append(code.text)
    for code in logt:
        logt_list.append(code.text)
    pIndex+=1
    
SchoolZone_info['LOC'] = loc_list
SchoolZone_info['CCTV_YN'] = cctv_yn_list
SchoolZone_info['CCTV_CNT'] = cctv_cnt_list
SchoolZone_info['ROAD_BT'] = road_bt_list
SchoolZone_info['LATITUDE'] = lat_list
SchoolZone_info['LONGITUDE'] = logt_list

df = pd.DataFrame(SchoolZone_info)
df = df[['LOC','CCTV_YN','CCTV_CNT','ROAD_BT','LATITUDE','LONGITUDE']]

df
```
